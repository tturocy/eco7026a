```python
# -*- coding: utf-8 -*-
"""
@author: Dr Andrea Calef 
"""

# Task: download from Yahoo Finance the daily time series of Dow Jones, Ftse 100 and the Mexican stock exchange index for the last year. 

#%reset -f 

import time
import datetime
import pandas as pd

# Add lines 15 and 16 in the case you need to change the folder directory. 
#import os 
#os.chdir("")

tickers = ['^DJI','^FTSE','^MXX']
# This example is with three time series, but you can modify by adding or cutting time series. 

period1 = int(time.mktime(datetime.datetime(2020,11,19).timetuple())) # initial period. 
period2 = int(time.mktime(datetime.datetime(2021,11,19).timetuple())) # final period. 
interval = '1d' # 1d as we wanted daily data. To have weekly or monthly data, you need to type '1wk' or '1m', respectively. 

# time.mktime is a method of Time module is used to convert a time.struct_time object or a tuple containing 9 elements corresponding to time.struct_time object to time in seconds passed since epoch in local time. 
# On Windows and most Unix systems, the epoch is January 1, 1970, 00:00:00 (UTC) and leap seconds are not counted towards the time in seconds since the epoch. To check what the epoch is on a given platform we can use time.gmtime(0). 

xlwriter = pd.ExcelWriter('Historical prices.xlsx', engine='openpyxl') 

for i in tickers: 
    query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{i}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'
    df = pd.read_csv(query_string) # it creates a DataFrame on Python. 
    df['ticker']=i
    df.to_excel(xlwriter,sheet_name=i, index=False) 
    # Line 26's code transforms the DataFrance for an Excel file. 
    # Note that every ticker's time series is saved in a different Excel worksheet. 
    
xlwriter.save()
# print(df) # only if you have a single ticker 



# To import from different sheets of the same Excel file and append them one below the other one.

#%reset -f 

# As we reset in line 43, we need to import pandas and repeat line 18 in line 53. 

import pandas as pd

workbook = pd.ExcelFile('Historical prices.xlsx')

sheets = workbook.sheet_names


df = pd.concat(pd.read_excel(workbook, sheet_name = None),  ignore_index=True)

print(df)

df.to_excel('Historical prices appended.xlsx')
df.to_stata('Historical prices appended.dta')

```
